## Подготовка обучающей и тестовой выборки

```python
df = df[df["MAIN_TARGET"].isin(["A", "B"])]
```

```python
df["TARGET"] = df["MAIN_TARGET"].apply(lambda x: 0 if x == "A" else 1)
```

```python
HOLD = df[df["MONTH_KEY"].isin(["2021-01-01", "2021-02-01"])]
```

```python
TRAIN = df[~df["MONTH_KEY"].isin(["2021-01-01", "2021-02-01"])]
```

```python
def get_X_y(data):
    data = data.drop(["SUBS_KEY", "BAN_KEY", "rank", "RTC_ACTIVE_IND", "TAC", "MONTH_KEY", "MAIN_TARGET"], axis=1)
    X, y = data.drop(["TARGET"], axis=1), data["TARGET"]
    return X, y
```

```python
X, y = get_X_y(TRAIN)
```

```python
features = list(X.columns)
```

```python
bool_cols = []
for col in features:
    if X[col].dtype == bool:
        print(col)
        bool_cols.append(col)
```

```python
def bool_to_float(df):
    X_ = df.copy()
    for i in bool_cols:
        X_[i] = X_[i].apply(lambda x: float(x))
    return X_
```

### Feature level constrains

```python
auto_woe = AutoWoE()
auto_woe = ReportDeco(auto_woe)
```

```python
auto_woe.fit(
    TRAIN[features + ['TARGET']].pipe(bool_to_float),
    target_name="TARGET",
    validation=HOLD.pipe(bool_to_float)[features + ['TARGET']]
)
```

```python
pred = auto_woe.predict_proba(HOLD)
roc_auc_score(HOLD['TARGET'], pred)
```

```python
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score, classification_report, roc_auc_score, precision_recall_curve
```

```python
def get_optimal_threshold(model, df, step=0.001, booster=False):
    from sklearn.metrics import f1_score
    import numpy as np
    def to_labels(pos_probs, threshold):
        return (pos_probs >= threshold).astype('int')
    thresholds = np.arange(0, 1, step)
    y = df["TARGET"]
    probs = model.predict_proba(HOLD)
    scores = [f1_score(y, to_labels(probs, t)) for t in thresholds]
    ix = np.argmax(scores)
    print('Threshold=%.3f, F-Score=%.5f' % (thresholds[ix], scores[ix]))
    return thresholds[ix], scores[ix]
```

```python
threshold, _ = get_optimal_threshold(auto_woe, HOLD)
```

```python
def save_model(pathmodel, model):
    import pickle

    with open(pathmodel, "wb") as fout:
        pickle.dump(model, fout)
```
